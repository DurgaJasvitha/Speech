```python
#A1. Load the recorded speech file into your python workspace. Once loaded, plot the graph for the 
#speech signal. You may use the below code from librosa as a reference.

import librosa
import librosa.display
import IPython.display as ipd

from itertools import cycle
y, sr = librosa.load(r"C:\Users\jasvi\Downloads\speechsignal.wav")
librosa.display.waveshow(y)
```

```python
# A2. Observe the length & magnitude range of the signal from the plot. Observing the plot, relate to
#the spoken words and silence between the words.

print("Sampling rate of the signal:",sr)
print("The speech signal y:", y)
print("Shape of the signal", y.shape)


#Observation: Magnitude decreases to zero when there is a silence in the plot. Increase to max when the words are spoken
```

```python
#trimming the signal
y_trimmed, _ = librosa.effects.trim(y, top_db=20) #the frequencies below 20db are trimmed
librosa.display.waveshow(y_trimmed)
```

```python
from IPython.display import Audio
Audio(data=y_trimmed, rate=sr) #playing the trimmed audio
```
