A4. Now you have acquainted yourself with spectral amplitudes of various consonants and vowel-basedphonemes. Generate the spectrogram of thesignal and observe the change points of the signals with associatedspeech segments. Observe to identify the consonants and vowels from the spectrogram.

```python
#vowels spectrogram

hop_size = int(0.02 *sr_vowels / 2)
window_length=int(sr_vowels*0.02) # windows of size 20ms
stft_matrix = librosa.stft(y_vowels, n_fft=window_length, hop_length=window_length)
# Convert complex spectrogram to magnitude spectrogram
magnitude_stft = np.abs(stft_matrix)
# Plot the magnitude spectrogram as a heatmap
plt.figure(figsize=(10, 6))
librosa.display.specshow(librosa.amplitude_to_db(magnitude_stft, ref=np.max), sr=sr_vowels, hop_length=window_length, x_axis='time', y_axis='hz')
plt.colorbar(format='%+2.0f dB')
plt.title('Magnitude Spectrogram of vowel speech Signal')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.tight_layout()
plt.show()
```

```python
#consonents pectrogram

hop_size = int(0.02 *sr_consonents/ 2)
window_length=int(sr_consonents*0.02) # windows of size 20ms
stft_matrix1 = librosa.stft(y_consonents, n_fft=window_length, hop_length=window_length)
# Convert complex spectrogram to magnitude spectrogram
magnitude_stft1 = np.abs(stft_matrix1)
# Plot the magnitude spectrogram as a heatmap
plt.figure(figsize=(10, 5))
librosa.display.specshow(librosa.amplitude_to_db(magnitude_stft1, ref=np.max), sr=sr_consonents, hop_length=window_length, x_axis='time', y_axis='hz')
plt.colorbar(format='%+2.0f dB')
plt.title('Magnitude Spectrogram of Consonent speech Signal')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.tight_layout()
plt.show()
```
